```python
# ============================================================
# IA PARA ENTRENADOR DEPORTIVO
# Uso de if, elif, else y while True
# Nivel: Principiante
# CBTIS 128
# Alumno: Samuel Nicola
# Inteligencia Artificial
# ============================================================

print("======================================")
print("🤖 IA PARA ENTRENADOR Y JUGADOR 🤖")
print("======================================")

while True:

    print("\n--------- MENÚ ---------")
    print("1. Evaluar jugador")
    print("2. Recomendar entrenamiento")
    print("3. Salir")

    opcion = input("Selecciona una opción: ")

    # OPCIÓN 1
    if opcion == "1":

        print("\n--- EVALUACIÓN DEL JUGADOR ---")

        cansancio = int(input("Nivel de cansancio del jugador (1-10): "))

        if cansancio >= 8:
            print("⚠️ El jugador está muy cansado.")
            print("Debe descansar.")

        elif cansancio >= 5:
            print("😐 El jugador tiene cansancio moderado.")
            print("Puede hacer entrenamiento ligero.")

        else:
            print("💪 El jugador está listo para entrenar fuerte.")

    # OPCIÓN 2
    elif opcion == "2":

        print("\n--- RECOMENDACIÓN DE ENTRENAMIENTO ---")

        objetivo = input("¿Qué deseas mejorar? (velocidad/fuerza/resistencia): ")

        if objetivo == "velocidad":
            print("🏃 Realizar ejercicios de sprint.")

        elif objetivo == "fuerza":
            print("🏋️ Realizar ejercicios con pesas.")

        elif objetivo == "resistencia":
            print("🔥 Realizar ejercicios de cardio.")

        else:
            print("❌ Opción no válida.")

    # OPCIÓN 3
    elif opcion == "3":

        print("\n👋 Gracias por usar la IA del entrenador.")
        print("Programa finalizado por Samuel Nicola NL = 44.")
        break

    # OPCIÓN INCORRECTA
    else:
        print("❌ Opción inválida.")
```
